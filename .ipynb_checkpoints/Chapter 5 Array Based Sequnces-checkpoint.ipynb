```python
# R-5.4

"""
Our DynamicArray class, as given in Code Fragment 5.3, does not support
use of negative indices with getitem . Update that method to better
match the semantics of a Python list.
"""

import ctypes

class DynamicList:
    def __init__(self):
        """Create an empty array"""
        self._n = 0
        self._capacity = 1
        self._A = self._make_array(self._capacity)
    def _resize(self, c):
        B = self._make_array(c)
        for i in B:
            B[i] = self._A[i]
        self._A = B
        self._capacity = c
    def append(self, element):
        if self._capacity == self._n:
            self._resize(2 * self._capacity)
        else:
            self._A[self._n] = element
            self._n += 1
    def __getitem__(self, num):
        if 0 <= num <= self._n:
            return self._A[num]
        elif num < 0:
            return self._A[self._n - num]
        else:
            raise IndexError("Invalid Index")
    def __len__(self):
        return self._n
    def _make_array(self, c):
        return (c*ctypes.py_object)()
    def insert(self,k,c):
        '''Here k < n.'''
        if self._n == self._capacity:
            self._resize(2 * self._capacity)
        for i in range(self._n, k, -1):
            self._A[i+1] = self._A[i]
        self._A[k] = c
        self._n += 1
    def remove(self,c):
        for i in range(self._n):
            if self._A[i] == c:
                for j in range(i, self._n-1):
                    self._A[j] = self._A[j+1]
                self._A[self._n-1] = None
                self._n -= 1
                return 
        raise ValueError('value not found')
```

```python
# R-5.6 

import ctypes

class DynamicList:
    def __init__(self):
        """Create an empty array"""
        self._n = 0
        self._capacity = 1
        self._A = self._make_array(self._capacity)
    def _resize(self, c):
        B = self._make_array(c)
        for i in range(self._n):
            B[i] = self._A[i]
        self._A = B
        self._capacity = c
    def _insert_resize(self, c, k):
        """k is the index to insert at"""
        B = self._make_array(c)
        for i in range(self._n):
            if i != k:
                """If i!=k continue as normal"""
                B[i] = self._A[i]
            else:
                """If i==k put a None"""
                B[k] = None
        self._A = B
        self._capacity = c
    def append(self, element):
        if self._capacity == self._n:
            self._resize(2 * self._capacity)
        else:
            self._A[self._n] = element
            self._n += 1
    def __getitem__(self, num):
        if 0 <= num <= self._n:
            return self._A[num]
        else:
            raise IndexError("Invalid Index")
    def __len__(self):
        return self._n
    def _make_array(self, c):
        return (c*ctypes.py_object)()
    def insert(self,k,c):
        '''Here k < n.'''
        if self._n == self._capacity:
            self._insert_resize(2 * self._capacity)
        self._A[k] = c
        self._n += 1
    def remove(self,c):
        for i in range(self._n):
            if self._A[i] == c:
                for j in range(i, self._n-1):
                    self._A[j] = self._A[j+1]
                self._A[self._n-1] = None
                self._n -= 1
                return 
        raise ValueError('value not found')
```

```python
# R-5.11 

l1 = [[i for i in range(5)] for j in range(5)]
l2 = [[1,2,3],
     [4,5,6],
     [7,8,9]]


# Approach 1: Use pointers to go through each element of the nxn list
# Time taken is O(n^2) as 2 for loops are used, space is O(1)

def approach1(l):
    s = 0
    for i in range(len(l)):
        for j in range(len(l[i])):
            s += l[i][j]
    return s

print(approach1(l1))
print(approach1(l2))


# Approach 2: Use in-built sum function
# Time taken is O(n^2) as sum is used within for loop, space is O(1)

def approach2(l):
    s = 0
    for i in l:
        s += sum(i)
    return s

print(approach2(l1))
print(approach2(l2))
```
