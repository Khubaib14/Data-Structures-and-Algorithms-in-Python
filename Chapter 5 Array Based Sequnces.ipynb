# Array Based Sequences

## Reinforcement

```python
# R-5.1 

import sys 
data = []

n = 100

for i in range(n):
    a = len(data)
    b = sys.getsizeof(data)
    print(f"Length of Array: {a} Size in bytes: {b}")
    data.append(None)
    
    
"""
- An empty list requires 56 bytes of memory. This is used up by some state variables for the list.  
- The very first element leads to a jump of 32 bytes. This is in line with theory. This goes on for 4 elements.
- On fifth element, there is again a jump of 32 bytes. The array can now store 8 integers.
"""
```

```python
# R-5.2

import sys

data = []
current, prev = 0,0
n = 100

print("Len", "Bytes")

for i in range(n):
    a = sys.getsizeof(data)
    b = len(data)
    current = a
    if current != prev:
        print(b, current)
    prev = current
    data.append(None)
```

```python
# R-5.3

import sys

n = 100
data = [0] * n

for i in range(n):
    a = len(data)
    b = sys.getsizeof(data)
    print(f"Length of Array: {a} Size in bytes: {b}")
    data.pop()
```
